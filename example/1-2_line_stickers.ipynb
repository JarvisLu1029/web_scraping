```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://store.line.me/stickershop/product/29386925/zh-Hant'

response = requests.get(url)

soup = bs(response.text, 'lxml')

# print(soup)
```

```python
import json
# 使用 CSS 選擇器抓取所有符合條件的 li 元素
li_elements = soup.select("ul.mdCMN09Ul.FnStickerList > li.mdCMN09Li.FnStickerPreviewItem.static-sticker")

# 檢查抓取到的 li 元素
# print(li_elements[0])
print(li_elements[0]['data-preview'])

# 可以使用 len() 印出抓取到的 li 元素數量，確認是否有抓取到所有圖片的元素
print(f"{len(li_elements)} 個貼圖元素")
```

```python
import json, os

# 建立資料夾
folder_path = 'line_stickers'
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# 建立一個空的 list，準備放置貼圖資訊
list_line_stickers = []
for li in li_elements:
    # 取得整個 data-preview 的值
    str_data = li['data-preview'] # 另一種寫法：li.get("data-preview")

    # 把屬性的值(字串)轉成物件 .loads()
    obj = json.loads(str_data)

    # 將重要資訊放置在 list 當中，幫助我們後續進行資料下載與儲存
    list_line_stickers.append(obj)

```

```python
n = 0
# 使用迴圈下載每一張貼圖
for obj in list_line_stickers: 
    n += 1
    # 下載圖片 要動圖改成'animationUrl'

    # print(f"貼圖ID: {obj['id']}, 下載連結: {obj['staticUrl']}")
    image_result = requests.get(obj['staticUrl'])
    with open(f'./line_stickers/test-{n}.png', 'wb') as file:
        file.write(image_result.content)
```
